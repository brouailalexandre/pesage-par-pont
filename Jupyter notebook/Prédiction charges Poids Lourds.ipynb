#### Cas particuliers

```python
# Suppression des lignes 'axle groups' avec virgule
normandie = normandie[~normandie['axle groups'].apply(lambda x: '.' in x)]
millau = millau[~millau['axle groups'].apply(lambda x: '.' in x)]
```

#### Normandie

```python
normandie_df = pd.DataFrame()
normandie_df = pd.concat([normandie_df, timestamp_dummies(normandie['timestamp'])], axis=1)
normandie_df = pd.concat([normandie_df, warningFlags_dummies(normandie['warning flags'])], axis=1)
normandie_df[['offset', 'lane', 'N', 'subclass id']] = normandie[['offset', 'lane', 'N', 'subclass id']]
normandie_df['v'] = ms_kmh(normandie['v'])
normandie_df = pd.concat([normandie_df, axleGroups_dummies(normandie['axle groups'])], axis=1)
normandie_df['gross weight'] = kN_t(normandie['gross weight'])
normandie_df[['_long', 'T', 'X2R', 'sum_strip_1', 'sum_strip_2']] = normandie[['_long', 'T', 'X2R', 'sum_strip_1', 'sum_strip_2']]
normandie_df[list_multiple_var('load_', 1, 16)] = kN_t(normandie[list_multiple_var('load_', 1, 16)])
normandie_df[list_multiple_var('d_', 1, 15)] = normandie[list_multiple_var('d_', 1, 15)]
```

#### Millau 

```python
millau_df = pd.DataFrame()
millau_df = pd.concat([millau_df, timestamp_dummies(millau['timestamp'])], axis=1)
millau_df = pd.concat([millau_df, warningFlags_dummies(millau['warning flags'])], axis=1)
millau_df[['offset', 'lane', 'N', 'subclass id']] = millau[['offset', 'lane', 'N', 'subclass id']]
millau_df['v'] = ms_kmh(millau['v'])
millau_df = pd.concat([millau_df, axleGroups_dummies(millau['axle groups'])], axis=1)
millau_df['gross weight'] = kN_t(millau['gross weight'])
millau_df[['_long', 'T', 'X2R']] = millau[['_long', 'T', 'X2R']]
millau_df[list_multiple_var('load_', 1, 16)] = kN_t(millau[list_multiple_var('load_', 1, 16)])
millau_df[list_multiple_var('d_', 1, 15)] = millau[list_multiple_var('d_', 1, 15)]
```

#### Suppression outliers

```python
list_df = [normandie_df, millau_df]

for df in list_df:
    
    # Warning flags
    ## W0 == 1 (Multiple truck presence) et W1 == 8 (Neg. Axle load)
    df.drop(df[(df['W0'] == 1) | (df['W1'] == 8)].index, axis=0, inplace=True)
    
    # Vitesse
    ## v > 85 km/h (normandie) et v > 105 km/h (millau)
    if str(df) == str(list_df[0]):
        df.drop(df[df['v'] > 85].index, axis=0, inplace=True)
    else:
        df.drop(df[df['v'] > 105].index, axis=0, inplace=True)
        
    # Nombre d'axes
    ## N > 8
    df.drop(df[df['N'] > 8].index, axis=0, inplace=True)
    
    # Poids total
    ## 70t > gross weight > 1.5t et gross weight == 0t
    df.drop(df[(df['gross weight'] < 1.5) | (df['gross weight'] > 70) | (df['gross weight'] == 0)].index, axis=0, inplace=True)
    
    # Longueur totale
    ## 30 > _long >= 2
    df.drop(df[(df['_long'] <= 2) | (df['_long'] > 30)].index, axis=0, inplace=True)
    
    # Suppression colonnes pas de sens
    for i in list_multiple_var('d_', 8, 15):
        df.drop(i, axis=1, inplace=True)
        
    for i in list_multiple_var('load_', 9, 16):
        df.drop(i, axis=1, inplace=True)
    
    # Distance entre essieux
    ## 9 > d_x > 0.8
    for i in list_multiple_var('d_', 1, 7):
        df.drop(df[(df[i] < .8) | (df[i] > 9)].index, axis=0, inplace=True)
        
    # Charge sur essieu
    ## 20 > load_x > 1
    for i in list_multiple_var('load_', 1, 8):
        df.drop(df[(df[i] < 1) | (df[i] > 20)].index, axis=0, inplace=True)
```
